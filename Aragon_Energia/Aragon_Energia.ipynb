```python
# Funci√≥n para contar certificados por categor√≠a energ√©tica


def contar_clasificaciones(dataframe):
    """
    Cuenta cu√°ntos certificados hay por clase energ√©tica
    tanto en emisiones como en consumo.
    Devuelve dos DataFrames ordenados de mayor a menor frecuencia.
    """
    conteo_emisiones = dataframe['clasificacion_emisiones'].value_counts().sort_values(ascending=False)
    conteo_consumo = dataframe['clasificacion_consumo'].value_counts().sort_values(ascending=False)


    df_emisiones = conteo_emisiones.to_frame(name='cantidad')
    df_consumo = conteo_consumo.to_frame(name='cantidad')

    return df_emisiones, df_consumo
```

```python
# Crear una funci√≥n que clasifique las provincias o los municipios en funci√≥n de su eficiencia energ√©tica, utilizando un mapa donde cada letra del certificado corresponde a un valor num√©rico.

def valorarcion_energetica_provincia(dataframe,groupby_campo):   #groupby_campo (municipio o provincia)
    
    dics_cambio_valores = {'A': 7,'B': 6,'C': 5,'D': 4,'E':3,'F': 2,'G': 1,'-':0} 

    dataframe_c = dataframe.copy()       
   
    dataframe_c['clasificacion_emisiones'] = dataframe_c['clasificacion_emisiones'].replace(dics_cambio_valores)
    dataframe_c['clasificacion_consumo'] = dataframe_c['clasificacion_consumo'].replace(dics_cambio_valores)
    
    dataframe_gb = round(dataframe_c[[groupby_campo,'clasificacion_emisiones','clasificacion_consumo']].groupby(groupby_campo).mean(),2)

    dataframe_gb['media_clasificacion_final'] = round((dataframe_gb['clasificacion_emisiones'] + dataframe_gb['clasificacion_consumo'])/2,2)

    dataframe_gb = dataframe_gb.sort_values('media_clasificacion_final',ascending = False)

    return dataframe_gb
```

```python
# Funci√≥n para identificar las provincias o los municipios que generan mayor consumo el√©ctrico y emisiones de CO‚ÇÇ.

def max_cons_emis(dataframe,groupby_campo,numero_resultados):    #groupby_campo (municipio o provincia)

    emis = dataframe.groupby(groupby_campo)["emisiones_anual_kgCO2_m2"].max().reset_index(name = 'max_emisiones')
    cons = dataframe.groupby(groupby_campo)["consumo_anual_kWh_m2"].max().reset_index(name = 'max_consumo')

    emis_ordenado = emis.sort_values("max_emisiones",ascending = False).head(numero_resultados)
    cons_ordenado = cons.sort_values("max_consumo",ascending = False).head(numero_resultados)

    return emis_ordenado,cons_ordenado,numero_resultados
```

```python
# Funci√≥n para detectar OUTLIER extremos:

def detectar_valores_extremos(df, columna_valor, columna_tipo="tipo_edificio"):
    """
    Detecta los certificados con valores EXTREMOS seg√∫n el tipo de edificio.
    Se consideran extremos los valores mayores que:
    
        Q3 + 1.5 * IQR
    
    Devuelve un √∫nico DataFrame con:
    - id_certificado
    - tipo_edificio
    - valor extremo
    """

    lista_outliers = []

    for tipo in df[columna_tipo].unique():
        sub_grupo = df[df[columna_tipo] == tipo][columna_valor]

        if sub_grupo.empty:
            continue

        Q1 = sub_grupo.quantile(0.25)
        Q3 = sub_grupo.quantile(0.75)
        IQR = Q3 - Q1

        limite = Q3 + 1.5 * IQR

        outliers_tipo = df[(df[columna_tipo] == tipo) & (df[columna_valor] > limite)][["id_certificado", columna_tipo, columna_valor]]

        lista_outliers.append(outliers_tipo)

    if lista_outliers:
        return pd.concat(lista_outliers, ignore_index=True)
    else:
        return pd.DataFrame(columns=["id_certificado", columna_tipo, columna_valor])
```

```python
# Crear una funci√≥n que agrupe las construcciones por d√©cadas y detecte cu√°les tuvieron mayor actividad edilicia

def decade_boom_edilizio(dataframe, anio_inicio, anio_fin):
    """
    Cuenta cu√°ntos edificios fueron construidos en cada d√©cada dentro del rango indicado.
    El rango se adapta autom√°ticamente a cualquier siglo.
    """
    df_filtrado = dataframe[dataframe['anio_construccion'].between(anio_inicio, anio_fin)].copy()
   
    inicio_decada = (anio_inicio // 10) * 10
    fin_decada = ((anio_fin // 10) + 1) * 10

    lista_limites = list(range(inicio_decada, fin_decada + 10, 10))

    etiquetas = [f"{i}-{i+9}" for i in range(inicio_decada, fin_decada, 10)]

    df_filtrado["decada"] = pd.cut(df_filtrado["anio_construccion"],bins=lista_limites,labels=etiquetas,right=False)
 
    conteo_decadas = df_filtrado["decada"].value_counts().sort_index()

    return conteo_decadas
```

```python
# Crear el gr√°fico de barras para la visualizaci√≥n:

conteo_decadas = decade_boom_edilizio(Aragon_energia_copy, 1900, 2020)


plt.figure(figsize=(12,6))
plt.bar(conteo_decadas.index, conteo_decadas.values, color="#4C72B0")

plt.title("Actividad edilicia por d√©cada")
plt.xlabel("D√©cada")
plt.ylabel("N√∫mero de edificios construidos")


for i, v in enumerate(conteo_decadas.values):
    plt.text(i, v + (v*0.02), str(v), ha='center', fontsize=10)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
